### Diseña un algoritmo para resolver el problema por fuerza bruta

Respuesta

```python
from itertools import combinations, permutations, groupby # importamos las librerias requeridas
import math
```

```python
#. Paso 1.
A=list("123456789") # Lista de números
B= list("+-*/")     #Lista de operadores


P_A=list(permutations(A, 5))  # Permutaciones de cada lista
P_B=list(permutations(B, 4))

LP_A = [list(i) for i in P_A] # Lista de listas
LP_B = [list(i) for i in P_B]

print(len(LP_A)) # Número total de permutaciones numericas
print(len(LP_B)) # Número total de permutaciones de los operadores
print( LP_A[0])# primera permutación de la lista 
print( LP_B[2]) # tercer permutación de la lista 
```

```python
# Paso 2.


def motor(x , y):
    lista_A=0
    lista_B = ""
    suma_listas = len(x) + len(y)
    union_1 = groupby(((x[len(x)*i//suma_listas], y[len(y)*i//suma_listas]) for i in range(suma_listas)), key=lambda x:x[0])
    union_2=[j[i] for k,g in union_1 for i,j in enumerate(g)]
    union_final = " ".join(union_2)
    motor_operador = (eval(union_final))
    decimales, enteros = math.modf(motor_operador)
    
    if decimales == 0:
      lista_A = motor_operador
      lista_B = union_final
      
    return lista_B, lista_A
```

```python
# Paso 3.
itera=0
numero_it=0
resultado = int(input("Ingrese el número requerido: "))
print('número requerido:', resultado)
for i in LP_A:
  for j in LP_B:
    itera = itera + 1
    resultado_motor=motor(i,j)
        
    if resultado_motor[1] == resultado:
      decide = True
      numero_it = numero_it+1
      print( resultado_motor[0], '  =  ', resultado_motor[1],'Iteracion #:', itera ) 
        
    else:
      decide = False

if numero_it > 0:
  print('solución')
else:
  print('no hay soluciones')
print('Total soluciones:', numero_it)
```
